## Token Info


```python
endpoint = f"/getAddressInfo/{address}?apiKey=freekey"
url = base_url + endpoint
response = requests.get(url)

if response.status_code == 200:
    token_info_response = json.loads(response.text)
token_info_response
```

### Storing tokens in Dataframe

```python
token_info_columns = list(token_list[0]["tokenInfo"].keys())
token_info_columns
```

```python
tkn_column = token_info_columns+token_columns[1:]
# tkn_column.sort()
tkn_column
```

```python
import pandas as pd

# create an Empty DataFrame object
df = pd.DataFrame()
```

```python
token_list[2]['tokenInfo']
```

```python
flag = True
for token in token_list:
    token_info = pd.DataFrame([token["tokenInfo"]])
    token_ = token.copy()
    token_.pop("tokenInfo")
    balance_info = pd.DataFrame([token_])
    info = pd.concat([token_info, balance_info],axis=1)
    # Run once
    if flag:
        columns = list(info.columns)
        df = pd.DataFrame(columns=columns)
        flag = False
    df = pd.concat([df, info], axis=0)
```
