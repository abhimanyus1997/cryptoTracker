```python
import requests
from bs4 import BeautifulSoup
```

## Getting Request

```python
account = input()

if account.strip() == "":  # âœ… prevents empty input (including whitespace characters)
    account = "0xd8dA6BF26964aF9D7eEd9e03E53415D37aA96045"

print(account)
```

```python
import requests

# Making a GET request
account_url = "https://etherscan.io/address/"+account

# Added Useragent to bypass you are not bot
r = requests.get(account_url,
                 headers={'User-Agent': "Mozilla/5.0 (Windows NT 10.0 ; Win64 ;x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/109.0.0.0 Safari/537.36"}
    )

# check status code for response received
# success code - 200
print(r)

# print content of request
#print(r.content)
```

## Parsing HTML using BS

```python
# Parsing the HTML
soup = BeautifulSoup(r.content, 'html.parser')
print(soup.prettify())
```

```python
# Finding Ether Balance & usd bal
balance = soup.find_all('div', class_='col-md-8')
print(balance[0].text)  # ETH tokens
print(balance[1].text)  # Current Value
```
